## Finding the number of books published after 2000-01-01

```python
query = '''SELECT * 
FROM 
     books
WHERE 
    publication_date >= '2000-01-01' '''
pd.io.sql.read_sql(query, con = engine) 
```

As we can clearly see, more than 3/4 of all books were printed after 2000-01-01. We can assume that the app is going to specialize in relatively new books. 

```python
query = '''SELECT 
	COUNT(title) AS count
FROM
	books 
 WHERE 
    publication_date >= '2000-01-01' '''
pd.io.sql.read_sql(query, con = engine) #we can see here that there are 821 books published after 2000-01-01
```

## Finding the average rating by book and the number of user reviews per book.

```python
query = ''' SELECT
    books.book_id AS id,
    COUNT(DISTINCT reviews.review_id) AS review_cnt,
    AVG(ratings.rating) AS avg_rating,
    books.title AS book_title
FROM
    books
    RIGHT JOIN reviews ON reviews.book_id = books.book_id
    RIGHT JOIN ratings ON ratings.book_id = books.book_id
    GROUP BY id,book_title
    ORDER BY review_cnt DESC '''
pd.io.sql.read_sql(query, con = engine)  #here is average rating by book and the number of user reviews per book
```

There are not many books, only a thousand; however. as we can see, quite a few of them have a high rating and many reviews. This means, that the people making the app made an effort to gather books that attract positive attention.

## Finding the publisher that has released the greatest number of books with more than 50 pages.

```python
query = ''' SELECT 
books.publisher_id AS publisher_id, 
publishers.publisher AS publisher,
COUNT(book_id) AS cnt_books
FROM books 
LEFT JOIN publishers  ON publishers.publisher_id = books.publisher_id
WHERE books.num_pages> 50
GROUP BY books.publisher_id,publisher
ORDER BY cnt_books DESC
LIMIT 1'''

pd.io.sql.read_sql(query, con = engine)#Here is the publisher house with the maximum number of books
```

We can see the publisher house with the maximum number of books has 42 books, quite a few. This information on the publisher would allow people to search books by publishing house.

## Finding the author with the highest average book rating among the books with at least 50 ratings.

```python
query = ''' SELECT
  books.book_id,
  AVG(ratings.rating) AS avg_rating,
  COUNT(DISTINCT ratings.rating_id) AS rating_cnt,
  authors.author AS author
FROM books
LEFT JOIN ratings
  ON books.book_id = ratings.book_id
LEFT JOIN authors
  ON books.author_id = authors.author_id

GROUP BY author, books.book_id
HAVING COUNT(DISTINCT ratings.rating_id) >= 50
ORDER BY AVG(ratings.rating) DESC
LIMIT 1'''
pd.io.sql.read_sql(query, con = engine) 
```

We can see that the author with the highest average rating is, unsurprisingly, J.K. Rowling (Mary GrandPré is the illustrator). J.K. Rowling is a world famous author that keeps attracting new readers even now. Our app picks most popular books concentrating on quality over quantity.

## Finding the average number of text reviews among users who rated more than 50 books.

```python
query = '''SELECT
AVG(cnt_review)  AS avg_cnt_review
FROM
(SELECT
 ratings.username as user_name, 
  COUNT(DISTINCT reviews.text)  AS cnt_review,
  COUNT(DISTINCT ratings.rating_id) AS rating_cnt
FROM reviews
RIGHT JOIN ratings
  ON reviews.book_id = ratings.book_id AND
   reviews.username = ratings.username

GROUP BY user_name 
HAVING COUNT(DISTINCT ratings.rating_id) >= 50) AS avg_review
LIMIT 1'''

pd.io.sql.read_sql(query, con = engine)
```

The average number of text reviews is 24, which is a really high amount. This means, that the books the app suggests attract public interest. Even the books with low ratings apparently attract reviews, which means, people are interested in reading them and they would still be profitable for the app.

Overall we can see that the selection of books chosen for this app would attract many users and we can anticipate a high level of engagement from the readers judging by the number of reviews.
